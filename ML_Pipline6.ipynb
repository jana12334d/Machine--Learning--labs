#Macchine Learning Pipline
Read-->Split-->divide-->train-->predict-->evaluate

```python
import numpy as np
import pandas as pd

data ={
    "area":[100,150,200,250,300,],
    "bedrooms":[2,3,3,4,5],
    "age":[20,15,10,5,2],
    "price":[200,300,400,500,650]
}

df=pd.DataFrame(data)
df.to_csv("housing.csv",index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

#Step1 Read

```python
df=pd.read_csv("housing.csv")
df.head()
```

#Step2 Separate features and the label

```python
x= df[["area","bedrooms","age"]]
y=df["price"]
```

#Step3 Splite the data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

#Step 4Train modle
#step4.1 Create modle
#step4.2 Fit modle

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

#Step 5 Predict

```python
y_pred=model.predict(x_test)
```

#Step 6 Evaluate

```python
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print("RMSE",rmse)
print("MSE",mse)
r2=r2_score(y_test,y_pred)
print("R2",r2)
```

#ML Pipline for linear classification

```python
import numpy as np
import pandas as pd

data ={
    "hours_studied":[2,3,5,6,7,8,9,10], # Added '10' to make its length 8
    "attendance":[50,60,70,80,85,90,95,100],
    "passed":[0,0,0,1,1,1,1,1]
}

df=pd.DataFrame(data)
df.to_csv("Student_csv",index=False)
```

```python
df=pd.read_csv("Student_csv")
df.head()
```

```python
x= df[["hours_studied","attendance",]]
y=df["passed"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from numpy.random import logistic
model = logistic()
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1_score = f1_score(y_test, y_pred)
print ("Accuracy:", accuracy)
print("Precision:", precision)
print ("Recall:", recall)
print ("F1 Score:", f1_score)
```
